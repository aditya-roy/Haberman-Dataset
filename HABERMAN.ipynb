Univariate Analysis

```python
#Probability Density Function

#plotting PDF for column: Age, axil_nodes_det, Op_Year 
sns.set_style("whitegrid")
f, axes = plt.subplots(ncols=3,figsize=(14, 7))
sns.distplot( h["Age"] , color="red", ax=axes[0])
sns.distplot( h["axil_nodes_det"] , color="blue", ax=axes[1])
sns.distplot( h["Op_Year"] , color="green", ax=axes[2])


```

Observations:

1. Maximum patients are in the range of 50-60.
2. Maximum people have lymph nodes ranging from 0-5.
3. Maximum people have been operated from 1958-1968.


Cumilitive Distributive Function.

```python
f, axes = plt.subplots(ncols=3,figsize=(14, 7))

sns.distplot(h.Age,
             hist_kws=dict(cumulative=False),
             kde_kws=dict(cumulative=True),ax=axes[0],color='red')
sns.distplot(h.axil_nodes_det,
             hist_kws=dict(cumulative=False),
             kde_kws=dict(cumulative=True),ax=axes[1],color="blue")
sns.distplot(h.Op_Year,
             hist_kws=dict(cumulative=False),
             kde_kws=dict(cumulative=True),ax=axes[2],color="green")
```

OBSERVATIONS:

1. 13-14% OF THE PEOPLE HAVE AGE BELOW 40.
2. 75% OF THE PEOPLE HAVE THE AGE BELOW 60.
3. 88% PATIENTS HAVE LESS THAN 10 POSITIVE LYMPH NODE.
4. 99% PATIENTS HAVE LESS THAN 30 POSITIVE LYMPH NODE.
5. 80% PATIENTS WERE ENROLLED BEFORE 1966.
6. 25% PATIENTS WERE ENROLLED BEFORE 1960.

```python
#Box Plot and whiskers by taking y-axis as 'Age'
sns.boxplot(x='surv_status',y='Age', data=h)
plt.show()
```

OBSERVATIONS:

1.PATIENTS WHO ARE HAVING AGE 30 TO 35 HAVE SURVIVED.

2.PATIENTS WHO ARE HAVING AGE GREATER THAN 78 HAVE NOT SURVIVED.

3.THE AVERAGE AGE OF THE PATIENTS WHO HAVE SURVIVED AND THOSE WHO HAVE DIED ARE THE ALMOST SAME.

```python
#Box Plot and whiskers by taking y-axis as 'axil_nodes_det'
sns.boxplot(x='surv_status',y='axil_nodes_det', data=h)
plt.show()
```

OBSERVATIONS:

POSITIVE LYMPH NODE

1.PATIENTS WHO ARE HAVING POSITIVE LYMPH NODES HAVE DIED MORE.

```python
#Box Plot and whiskers by taking y-axis as 'Op_Year'
sns.boxplot(x='surv_status',y='Op_Year', data=h)
plt.show()
```

OBSERVATIONS:

YEAR OF OPERATION

1. PATIENTS WHO WERE OPERATED BEFORE 1960 HAVE LESS CHANCES OF SURVIVAL.

3. THE AVERAGE OPERATION YEAR OF THE PATIENTS WHO HAVE SURVIVED AND THOSE WHO HAVE DIED ARE THE ALMOST SAME(1963).
